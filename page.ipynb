```python
#mainly selenium code

# create a class for each element (textbox)
#that inherets from BasePageElement(locator.ipynb)
# so the setter and the getter will be called automatically 
# and the value will be set.



class BasePage:
    
    def __init__(self,driver):
        self.driver = driver
        


#create class for each page         
class RegisterPage(BasePage):
    
    # Initializing variables that will be set in (test.py) to the specified value
    first_name_textbox = FirstNameTextBox()
    last_name_textbox = LastNameTextBox()
    
    phone_textbox = PhoneTextBox()
    email_textbox = EmailTextBox()
    
    password_textbox = PasswordTextBox()
    confirm_password_textbox = ConfirmPasswordTextBox()
    
    
    
    #check if the current url matches with the url 
    def is_title_matches(self):
        return self.driver.current_url == utils.PHP_URL_REGISTER
    
        
    #check if the current url doesn't match with the url  
    def is_title_differs(self):
        return self.driver.current_url != utils.PHP_URL_REGISTER

    def click_signup_button(self):
        element = self.driver.find_element(*RegisterPageLocators.SIGN_UP_BUTTON)
        element.click()
        
    # check if each element is present in the page 
    def Check_Page_Elements(self):

        firstname = self.driver.find_element(*RegisterPageLocators.FIRST_NAME_TEXTBOX)
        check1 = firstname.is_displayed()

        lastname = self.driver.find_element(*RegisterPageLocators.LAST_NAME_TEXTBOX)
        check2 = lastname.is_displayed()

        phone = self.driver.find_element(*RegisterPageLocators.PHONE_TEXTBOX)
        check3 = phone.is_displayed()

        email = self.driver.find_element(*RegisterPageLocators.EMAIL_TEXTBOX)
        check4 = email.is_displayed()

        password = self.driver.find_element(*RegisterPageLocators.PASSWORD_TEXTBOX)
        check5 = password.is_displayed()

        confirm_password = self.driver.find_element(*RegisterPageLocators.CONFIRMEDPASSWORD_TEXTBOX)
        check6 = confirm_password.is_displayed()

        sign_up = self.driver.find_element(*RegisterPageLocators.SIGN_UP_BUTTON)
        check7 = sign_up.is_displayed()
        
        logo = self.driver.find_element(*RegisterPageLocators.LOGO)
        check8 = logo.is_displayed()     

        title = self.driver.find_element(*RegisterPageLocators.TITLE)
        check9 = title.is_displayed()
        
        return ( check1 and check2 and check3 and check4 and check5 and check6 and check7 and check8 and check9)
        
        
class AccountPage(BasePage):
    
    
    def is_title_matches(self):
        return self.driver.current_url == utils.PHP_URL_ACCOUNT
    
        
    #check if the current url doesn't match with the url  
    def is_title_differs(self):
        return self.driver.current_url != utils.PHP_URL_ACCOUNT


    def click_account_button(self):
        
        element = self.driver.find_element(*AccountPageLocators.DROPDOWN_BUTTON)
        element.click()
        
    def click_logout_button(self):
        
        element = self.driver.find_element(*AccountPageLocators.LOGOUT_BUTTON)
        element.click()
    
    
    
class LoginPage(BasePage):
    
    
    email_textbox = EmailTextBox_SignUp()
    
    password_textbox = PasswordTextBox_SignUp()
    
    #check if the current url matches with the url 
    def is_title_matches(self):
        return self.driver.current_url == utils.PHP_URL_LOGIN
    
        
    #check if the current url doesn't match with the url  
    def is_title_differs(self):
        return self.driver.current_url != utils.PHP_URL_LOGIN


    def click_login_button(self):
        
        element = self.driver.find_element(*LoginPageLocators.LOGIN_BUTTON)
        element.click()
        
    # check if each element is present in the page 
    def Check_Page_Elements(self):

        email = self.driver.find_element(*LoginPageLocators.EMAIL_TEXTBOX)
        check1 = email.is_displayed()

        password = self.driver.find_element(*LoginPageLocators.PASSWORD_TEXTBOX)
        check2 = password.is_displayed()

        login = self.driver.find_element(*LoginPageLocators.LOGIN_BUTTON)
        check3 = login.is_displayed()

        logo = self.driver.find_element(*LoginPageLocators.LOGO)
        check4 = logo.is_displayed()     

        title = self.driver.find_element(*LoginPageLocators.TITLE)
        check5 = title.is_displayed()

        return (check1 and check2 and check3 and check4 and check5 )
        
    
    

```
