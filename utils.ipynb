```python
import time 
from PIL import Image
import import_ipynb
import time
from selenium import webdriver
import os
import json
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.chrome.options import Options
import requests

PHP_URL_REGISTER = 'https://www.phptravels.net/register'

PHP_URL_ACCOUNT = "https://www.phptravels.net/account/"

PHP_URL_LOGIN = "https://www.phptravels.net/login"

CHROME_DRIVER = 'driver/chromedriver.exe'


# takes the data and fill the textboxes
def fill_register_data(register_page,first_name,last_name,phone,email,password,confirmpassword):
    
    register_page = register_page
    register_page.first_name_textbox = first_name
    time.sleep(1)
    
    register_page.last_name_textbox = last_name
    time.sleep(1)
    
    register_page.phone_textbox = phone
    time.sleep(1)
    
    register_page.email_textbox = email
    time.sleep(1)
    
    register_page.password_textbox = password
    time.sleep(1)
    
    register_page.confirm_password_textbox = confirmpassword
    time.sleep(1)
    
    register_page.click_signup_button()
    time.sleep(3)

    
def fill_login_data(login_page,email,password):
    
    login_page = login_page
    login_page.email_textbox = email
    time.sleep(2)
    
    login_page.password_textbox = password
    time.sleep(2)
    login_page.click_login_button()
    time.sleep(4)
    
def go_to_login_page(account_page):
    
    account_page.click_account_button()
    time.sleep(2)
    account_page.click_logout_button()
    time.sleep(2)
    
        

#creates a valid password with uppercases and lowercases
def valid_password(p):
    password = p[0:7]
    password1 = password[:len(password)//2].lower()
    password2 = password[len(password)//2:].upper()
    return password1 + password2


def take_a_screenshot(driver):
    
    if not os.path.exists("reports"):
        os.mkdir("reports")
        
    if not os.path.exists("reports/screenshots"):
    
        os.mkdir("reports/screenshots")
        
    folder ='reports/screenshots/'
    time_str = time.asctime().replace(':','')
    file_name = folder + time_str +'.png'
    driver.save_screenshot(file_name)
    resize_screenshot(file_name)
    #driver.get_screenshot_as_file(file_name)


def resize_screenshot(file):
    screenshot = Image.open(file)
    screenshot = screenshot.resize((1024,680)).convert('RGB')
    screenshot.save(file)

    
#options for the driver  
def options():
    options = Options()
    options.add_argument('start-maximized')
    prefs = {"profile.managed_default_content_settings.images": 2}
    options.add_experimental_option("prefs", prefs)
    return options

def email_wrong_order(email):
    email2 = email.split('@')
    email3 = email2[1].split('.')
    email4 = email2[0]+"@."+email3[1]
    return email4

def email_invalid2(email):
    email2 = email.split('@')
    email3 = email2[1].split('.')
    email4 = email2[0]+"."+email3[0]+"@"+email3[1]
    return email4


def check_status(url):
    r = requests.get(url)
    return r.status_code

def save_email(email):
    with open('email.txt', 'a') as file:
        file.writelines(email + '\n')
        file.close()


def get_email():
    file = open("email.txt", "r")
    return(file.readline())

def http_request(driver):
    if not os.path.exists("logs"):
        os.mkdir("logs")
    data = {}
    folder = "logs/"
    time_str = time.asctime().replace(':','')
    file_name = folder +time_str+".txt"
    with open(file_name, 'a') as file:
            for entry in driver.get_log('browser'):
                json.dump(entry,file)
                  
```
